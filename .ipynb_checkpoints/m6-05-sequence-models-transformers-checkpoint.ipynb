```python
# Core PyTorch libraries for building and training neural networks
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset

# Data manipulation and math utilities
import numpy as np
import pandas as pd
import re
from collections import Counter
from sklearn.model_selection import train_test_split

# Determine if a GPU is available, otherwise default to CPU
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

# Set a manual seed for reproducibility so we get the same results on every run
torch.manual_seed(42)
```

```python
# Try to load the dataset using the HuggingFace 'datasets' library
try:
    from datasets import load_dataset
    print("Loading IMDB dataset from HuggingFace...")
    dataset = load_dataset("imdb")
    
    # Convert the training split to a Pandas DataFrame
    df = pd.DataFrame(dataset['train'])
    
    # Randomly sample 5000 reviews for faster processing and reset the index
    df = df.sample(5000, random_state=42).reset_index(drop=True)
    
except ImportError:
    print("Please install datasets (`pip install datasets`) or load a local CSV fallback.")
    # Fallback example if using a local CSV:
    # df = pd.read_csv("imdb.csv").sample(5000, random_state=42).reset_index(drop=True)

# Extract the raw text reviews and their corresponding binary labels (0 = negative, 1 = positive)
texts = df['text'].tolist()
labels = df['label'].tolist()
```

```python
def tokenize(text):
    """
    Cleans and tokenizes a single text string.
    """
    # 1. Convert everything to lowercase so "The" and "the" are treated as the same word
    text = text.lower()
    
    # 2. IMDB reviews are scraped from the web and often contain HTML break tags. We replace them with spaces.
    text = re.sub(r'<br\s*/?>', ' ', text) 
    
    # 3. Remove basic punctuation (keeps alphanumeric characters and whitespace)
    text = re.sub(r'[^\w\s]', '', text)    
    
    # 4. Split the text into a list of words based on whitespace
    return text.split()

print("Tokenizing texts...")
# Apply the tokenize function to every review in our dataset
tokenized_texts = [tokenize(text) for text in texts]
```

```python
vocab_size = 10000 # We will only keep track of the top 10,000 words

# Count the frequency of every word across all tokenized reviews
counter = Counter()
for tokens in tokenized_texts:
    counter.update(tokens)

# Retrieve the most common words. 
# We subtract 2 from vocab_size to leave room for our special <pad> and <unk> tokens.
most_common = counter.most_common(vocab_size - 2)

# Initialize the vocabulary with special tokens:
# <pad> (index 0): Used to fill short sentences so all inputs are the same length.
# <unk> (index 1): "Unknown". Used when we encounter a word not in our top 10,000.
vocab = {"<pad>": 0, "<unk>": 1}

# Populate the rest of the vocabulary dictionary
# enumerate(..., start=2) ensures we start assigning integers from 2 onwards
for idx, (word, _) in enumerate(most_common, start=2):
    vocab[word] = idx
    
print(f"Vocabulary built with {len(vocab)} words.")
```

```python
MAX_LEN = 200 # The fixed length for every sequence

def encode_text(tokens, vocab, max_len):
    """
    Converts a list of string tokens into a list of integer IDs, 
    then pads or truncates it to the specified max_len.
    """
    # Convert words to IDs. If a word isn't in 'vocab', default to the <unk> token ID.
    encoded = [vocab.get(word, vocab["<unk>"]) for word in tokens]
    
    if len(encoded) < max_len:
        # If the review is too short, append the <pad> token ID (0) until it reaches max_len
        encoded = encoded + [vocab["<pad>"]] * (max_len - len(encoded))
    else:
        # If the review is too long, slice it to keep only the first max_len tokens
        encoded = encoded[:max_len]
        
    return encoded

# Apply encoding to all tokenized reviews
encoded_texts = [encode_text(tokens, vocab, MAX_LEN) for tokens in tokenized_texts]
```

```python
class IMDBDataset(Dataset):
    """
    Custom PyTorch Dataset for our encoded reviews and labels.
    """
    def __init__(self, texts, labels):
        # Convert lists to PyTorch tensors.
        # texts must be Long tensors because they are integer indices used for embeddings
        self.texts = torch.tensor(texts, dtype=torch.long)
        
        # labels must be Float tensors because binary classification loss functions 
        # (like BCEWithLogitsLoss) expect float inputs.
        self.labels = torch.tensor(labels, dtype=torch.float32)

    def __len__(self):
        # Returns the total number of samples
        return len(self.texts)

    def __getitem__(self, idx):
        # Fetches a single sample and its label at the specified index
        return self.texts[idx], self.labels[idx]

# Split the data: 80% for training, 20% for validation
X_train, X_val, y_train, y_val = train_test_split(
    encoded_texts, labels, test_size=0.2, random_state=42
)

# Instantiate the datasets
train_dataset = IMDBDataset(X_train, y_train)
val_dataset = IMDBDataset(X_val, y_val)

# Create DataLoaders to yield batches of 32 samples at a time
# We shuffle the training data to prevent the model from learning sequence-based patterns
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

print(f"Train batches: {len(train_loader)} | Val batches: {len(val_loader)}")
```

```python
# Find the shortest review in our tokenized set to clearly see the padding zeros at the end
lengths = [len(t) for t in tokenized_texts]
short_idx = np.argmin(lengths)

print("\n--- Sample Encoding Verification ---")
print("1. Original Text Snippet:")
print(f"{texts[short_idx][:100]}...") # Printing just the first 100 characters

print("\n2. Tokenized:")
print(tokenized_texts[short_idx])

print(f"\n3. Encoded Tensor (Length forced to: {len(encoded_texts[short_idx])}):")
print(encoded_texts[short_idx])
```

# Task 1 — Implement Scaled Dot-Product Attention from Scratch

Before using PyTorch's built-in attention, build your own.

1. Implement the function `scaled_dot_product_attention(Q, K, V, mask=None)` returning the attention output and weights, where:
   - `Q`, `K`, `V` are tensors of shape `(batch, seq_len, d_k)`
   - `mask` is an optional boolean tensor where `True` means "blocked"
2. Use it on a tiny example: 2 sequences of 4 tokens with `d_k=8`, all values random.
3. Verify your implementation matches `nn.functional.scaled_dot_product_attention` (or do a manual check that softmax rows sum to 1 and outputs match `weights @ V`).
4. Print the attention weight matrix for one example. In a markdown cell, comment on its shape and what each row represents.


```python
import math
import torch
import torch.nn.functional as F

def scaled_dot_product_attention(Q, K, V, mask=None):
    """
    Computes Scaled Dot-Product Attention.
    
    Args:
        Q (Tensor): Queries of shape (batch, seq_len, d_k)
        K (Tensor): Keys of shape (batch, seq_len, d_k)
        V (Tensor): Values of shape (batch, seq_len, d_v) -- usually d_v = d_k
        mask (Tensor, optional): Boolean mask where True means "blocked". 
                                 Shape can broadcast to (batch, seq_len, seq_len).
                                 
    Returns:
        output (Tensor): The context vectors (batch, seq_len, d_k)
        attn_weights (Tensor): The attention matrix (batch, seq_len, seq_len)
    """
    # 1. Extract d_k (the dimension of the keys/queries) for scaling
    d_k = Q.size(-1)
    
    # 2. Compute the dot product between Queries and Keys
    # We must transpose K on its last two dimensions to multiply (seq_len, d_k) x (d_k, seq_len)
    # Resulting shape: (batch, seq_len, seq_len)
    scores = torch.matmul(Q, K.transpose(-2, -1))
    
    # 3. Scale the scores to prevent the softmax gradients from vanishing
    scores = scores / math.sqrt(d_k)
    
    # 4. Apply the mask (if one is provided)
    if mask is not None:
        # The prompt states True means "blocked". 
        # We replace blocked positions with a huge negative number (-1e9) 
        # so that when softmax is applied, their probability becomes exactly 0.
        scores = scores.masked_fill(mask == True, -1e9)
        
    # 5. Apply Softmax along the last dimension (the keys) to get probabilities
    # Shape remains: (batch, seq_len, seq_len)
    attn_weights = F.softmax(scores, dim=-1)
    
    # 6. Multiply the attention weights by the Values
    # (batch, seq_len, seq_len) @ (batch, seq_len, d_k) -> (batch, seq_len, d_k)
    output = torch.matmul(attn_weights, V)
    
    return output, attn_weights
```

```python
# Set seed so we get the exact same random numbers every time
torch.manual_seed(42)

batch_size = 2
seq_len = 4
d_k = 8

# Create random tensors for Q, K, and V based on the required shapes
Q = torch.randn(batch_size, seq_len, d_k)
K = torch.randn(batch_size, seq_len, d_k)
V = torch.randn(batch_size, seq_len, d_k)

print(f"Query shape: {Q.shape}")
print(f"Key shape:   {K.shape}")
print(f"Value shape: {V.shape}")
```

```python
# 1. Run our custom implementation
custom_output, custom_weights = scaled_dot_product_attention(Q, K, V)

# 2. Run PyTorch's built-in implementation for comparison
# Note: F.scaled_dot_product_attention is available in PyTorch 2.0+
builtin_output = F.scaled_dot_product_attention(Q, K, V)

# 3. Verification Checks
print("--- VERIFICATION ---")
# torch.allclose checks if all elements are equal within a tiny floating-point tolerance
is_matching = torch.allclose(custom_output, builtin_output, atol=1e-6)
print(f"1. Matches nn.functional output exactly: {is_matching}")

# Check if softmax worked correctly (rows must sum to 1)
# We sum across the last dimension. We expect a tensor filled with 1.0s
row_sums = custom_weights.sum(dim=-1)
print(f"2. Row sums of attention weights (should be 1.0):\n{row_sums}")

# 4. Print the attention weight matrix for the first example in the batch
print("\n--- ATTENTION WEIGHT MATRIX (Example 0) ---")
# Extracting the 0th batch item. It will be a 4x4 matrix (seq_len x seq_len)
print(custom_weights[0])
```

### Analysis of the Attention Weight Matrix

**1. Matrix Shape:**
The attention weight matrix for a single example has a shape of `(4, 4)`, which represents `(seq_len, seq_len)`. Because our dummy sequence contains exactly 4 tokens, the matrix is a square 4x4 grid capturing the relationship between every possible pair of tokens.

**2. What Each Row Represents:**
* **The Rows (Queries):** Each row corresponds to a specific **Query** token (the word currently active and looking out at the rest of the sentence to gather context).
* **The Columns (Keys):** Each column in a row corresponds to a **Key** token (all the words in the sequence being evaluated for relevancy).
* **The Values (Attention Weights):** The values are probabilities between 0 and 1 that dictate how much focus the row's token places on the column's token. Because of the softmax layer, the values in any given row sum up to exactly `1.0`.

**Specific Examples from our Output Matrix:**
* **Row 0 (First Token):** The value at index `(0, 0)` is `0.5092`. This means the first token assigns **50.9%** of its attention to itself, while distributing the remaining focus to the other tokens (`31.6%`, `11.1%`, and `6.3%`).
* **Row 3 (Fourth Token):** The value at index `(3, 0)` is `0.7451`. This shows that the fourth token is paying a massive **74.5%** of its attention to the very first token in the sequence, demonstrating how the mechanism captures strong dependencies regardless of distance.

# Task 2 — LSTM Classifier

1. Define an `LSTMClassifier`:
   - `nn.Embedding(vocab_size, embed_dim=64)`
   - `nn.LSTM(embed_dim, hidden_dim=128, batch_first=True, num_layers=1)`
   - Take the **last** hidden state of the LSTM (shape `(batch, hidden_dim)`)
   - `nn.Dropout(0.3)` → `nn.Linear(hidden_dim, 2)` for binary classification
2. Train for **5 epochs** with `Adam(lr=1e-3)` and `CrossEntropyLoss`. Time the training with `time.time()`.
3. Plot training and validation loss + accuracy. Report best validation accuracy and total training time.


```python
import torch
import torch.nn as nn

class LSTMClassifier(nn.Module):
    def __init__(self, vocab_size, embed_dim=64, hidden_dim=128, output_dim=2):
        """
        Initialize the layers of the LSTM Classifier.
        """
        super(LSTMClassifier, self).__init__()
        
        # 1. Embedding Layer: Maps token indices to dense vectors of size embed_dim (64)
        self.embedding = nn.Embedding(vocab_size, embed_dim)
        
        # 2. LSTM Layer: Processes sequential embeddings. 
        # batch_first=True means input shape is (batch, seq_len, embed_dim)
        self.lstm = nn.LSTM(embed_dim, hidden_dim, batch_first=True, num_layers=1)
        
        # 3. Dropout Layer: Regularization technique to prevent overfitting by zeroing 30% of activations
        self.dropout = nn.Dropout(0.3)
        
        # 4. Linear Layer: Maps the final hidden state to 2 output classes (binary classification)
        self.fc = nn.Linear(hidden_dim, output_dim)

    def forward(self, text):
        """
        Defines the forward pass of the model.
        Input 'text' shape: [batch_size, seq_length]
        """
        # Step 1: Pass token IDs through embedding layer
        # Output shape: [batch_size, seq_length, embed_dim]
        embedded = self.embedding(text) 
        
        # Step 2: Pass embeddings into the LSTM layer
        # lstm_out shape: [batch_size, seq_length, hidden_dim]
        # hidden state shape: [num_layers, batch_size, hidden_dim]
        lstm_out, (hidden, cell) = self.lstm(embedded)
        
        # Step 3: Extract the last hidden state of the LSTM
        # hidden[-1] selects the state after processing the final token of the sequence
        # Output shape: [batch_size, hidden_dim]
        last_hidden = hidden[-1] 
        
        # Step 4: Apply dropout to the last hidden state
        dropped = self.dropout(last_hidden)
        
        # Step 5: Final linear layer to get raw logit predictions
        # Output shape: [batch_size, output_dim]
        output = self.fc(dropped) 
        
        return output
```

```python
# 1. Setup the computation device (Use GPU if available, otherwise fallback to CPU)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# 2. Instantiate the model using the vocab_size from your dataset
# vocab_size variable must be defined in your previous preprocessing cells
model = LSTMClassifier(vocab_size=vocab_size).to(device)

# 3. Define the loss function for binary/multi-class classification
criterion = nn.CrossEntropyLoss()

# 4. Define the Adam optimizer with the required learning rate of 1e-3 (0.001)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

# 5. Set hyperparameter for total epochs and arrays to track metrics for plotting
epochs = 5
train_losses, val_losses = [], []
train_accs, val_accs = [], []
best_val_acc = 0.0 # Track the highest accuracy achieved on validation set
```

```python
import time

print("Starting training pipeline...")
start_time = time.time()

for epoch in range(epochs):
    # --- TRAINING PHASE ---
    model.train()
    epoch_train_loss, correct_train, total_train = 0, 0, 0
    
    for texts, labels in train_loader:
        texts = texts.to(device)
        # FIX APPLIED: Cast labels to integer (long) to satisfy CrossEntropyLoss
        labels = labels.to(torch.long).to(device)
        
        optimizer.zero_grad()
        predictions = model(texts)
        loss = criterion(predictions, labels)
        
        loss.backward()
        optimizer.step()
        
        epoch_train_loss += loss.item()
        _, predicted = torch.max(predictions, 1)
        total_train += labels.size(0)
        correct_train += (predicted == labels).sum().item()
        
    train_losses.append(epoch_train_loss / len(train_loader))
    train_accs.append(correct_train / total_train)
    
    # --- VALIDATION PHASE ---
    model.eval()
    epoch_val_loss, correct_val, total_val = 0, 0, 0
    
    with torch.no_grad():
        for texts, labels in val_loader:
            texts = texts.to(device)
            # FIX APPLIED: Cast labels to integer (long) here as well
            labels = labels.to(torch.long).to(device)
            
            predictions = model(texts)
            loss = criterion(predictions, labels)
            
            epoch_val_loss += loss.item()
            _, predicted = torch.max(predictions, 1)
            total_val += labels.size(0)
            correct_val += (predicted == labels).sum().item()
            
    val_losses.append(epoch_val_loss / len(val_loader))
    val_acc = correct_val / total_val
    val_accs.append(val_acc)
    
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        
    print(f"Epoch {epoch+1}/{epochs} | "
          f"Train Loss: {train_losses[-1]:.4f}, Acc: {train_accs[-1]:.4f} | "
          f"Val Loss: {val_losses[-1]:.4f}, Acc: {val_accs[-1]:.4f}")

total_time = time.time() - start_time

print(f"\n--- Execution Report ---")
print(f"Total Training Time: {total_time:.2f} seconds")
print(f"Best Validation Accuracy: {best_val_acc * 100:.2f}%")
```

```python
import matplotlib.pyplot as plt

# ==========================================
# 4. PLOTTING RESULTS
# ==========================================
plt.figure(figsize=(14, 5))

# Plot 1: Loss curves
plt.subplot(1, 2, 1)
plt.plot(range(1, epochs+1), train_losses, label='Train Loss', marker='o', linewidth=2)
plt.plot(range(1, epochs+1), val_losses, label='Validation Loss', marker='o', linewidth=2)
plt.title('Training vs Validation Loss Performance')
plt.xlabel('Epochs')
plt.ylabel('Loss Value')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

# Plot 2: Accuracy curves
plt.subplot(1, 2, 2)
plt.plot(range(1, epochs+1), train_accs, label='Train Accuracy', marker='s', linewidth=2)
plt.plot(range(1, epochs+1), val_accs, label='Validation Accuracy', marker='s', linewidth=2)
plt.title('Training vs Validation Accuracy Performance')
plt.xlabel('Epochs')
plt.ylabel('Accuracy Percentage')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

# Task 3 — Transformer Classifier and Comparison with the LSTM

Build a transformer encoder for the same sentiment task and compare it head-to-head against the LSTM you trained in Task 2.

1. Define a `TransformerClassifier`:
   - `nn.Embedding(vocab_size, d_model=64)`
   - A learned positional encoding `nn.Embedding(max_len=200, d_model=64)`
   - `nn.TransformerEncoder` with `num_layers=2`, `d_model=64`, `nhead=4`, `dim_feedforward=128`, `batch_first=True`, `dropout=0.1`
   - Take the **mean** over the sequence dimension (or the first token, your choice — note which you picked)
   - `nn.Dropout(0.3)` → `nn.Linear(d_model, 2)`
2. Train for the same **5 epochs** with the same optimiser and loss. Time the training.
3. Plot training and validation loss + accuracy, report best validation accuracy and total training time, then fill in the comparison table below and write a **4–6 sentence comparison** of the two models — which converged faster per epoch, which ended at higher accuracy, and whether the transformer's parallelism noticeably affected training time on your hardware.

| Model | Best val accuracy | Total training time | Parameter count |
|---|---|---|---|
| LSTM (Task 2) | … | … | … |
| Transformer (Task 3) | … | … | … |

```python
import torch
import torch.nn as nn

class TransformerClassifier(nn.Module):
    def __init__(self, vocab_size, d_model=64, nhead=4, num_layers=2, dim_feedforward=128, max_len=200, output_dim=2):
        """
        Initialize the Transformer Classifier layers.
        """
        super(TransformerClassifier, self).__init__()
        
        # 1. Token Embedding: Converts token IDs into dense vectors of size d_model (64)
        self.token_embedding = nn.Embedding(vocab_size, d_model)
        
        # 2. Learned Positional Embedding: Learns position information up to max_len (200 tokens)
        self.pos_embedding = nn.Embedding(max_len, d_model)
        
        # 3. Transformer Encoder Layer: Represents a single encoder block
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model, 
            nhead=nhead, 
            dim_feedforward=dim_feedforward, 
            dropout=0.1, 
            batch_first=True # Keeps the shape as (batch, seq, feature)
        )
        
        # 4. Transformer Encoder: Stacks 'num_layers' (2) of encoder layers together
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        
        # 5. Dropout Layer: Regularization for the classification head
        self.dropout = nn.Dropout(0.3)
        
        # 6. Linear Layer: Maps the pooled output to 2 classes
        self.fc = nn.Linear(d_model, output_dim)

    def forward(self, text):
        """
        Input 'text' shape: [batch_size, seq_length]
        """
        batch_size, seq_len = text.size()
        
        # Step 1: Generate positions dynamically for the current batch sequence length
        # positions shape: [1, seq_length]
        positions = torch.arange(0, seq_len, device=text.device).unsqueeze(0)
        
        # Step 2: Combine Token Embeddings and Positional Embeddings
        x = self.token_embedding(text) + self.pos_embedding(positions) # [batch_size, seq_length, d_model]
        
        # Step 3: Pass through the stacked Transformer Encoder layers
        x = self.transformer_encoder(x) # [batch_size, seq_length, d_model]
        
        # NOTE ON POOLING CHOICE: We take the MEAN over the sequence dimension (dim=1) 
        # to compress the sentence into a single context vector.
        pooled_output = x.mean(dim=1) # [batch_size, d_model]
        
        # Step 4: Apply dropout and final linear classification layer
        dropped = self.dropout(pooled_output)
        output = self.fc(dropped) # [batch_size, output_dim]
        
        return output
```

```python
# 1. Setup the computation device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# 2. Instantiate the Transformer Classifier
# Note: vocab_size must be already defined in your notebook
transformer_model = TransformerClassifier(vocab_size=vocab_size).to(device)

# 3. Define the loss function (identical to LSTM setup)
criterion = nn.CrossEntropyLoss()

# 4. Define the Adam optimizer with a learning rate of 1e-3
optimizer = torch.optim.Adam(transformer_model.parameters(), lr=1e-3)

# 5. Hyperparameters and metrics tracking
epochs = 5
t_train_losses, t_val_losses = [], []
t_train_accs, t_val_accs = [], []
t_best_val_acc = 0.0
```

```python
import time

print("Starting Transformer training pipeline...")
t_start_time = time.time() # Start stopwatch for Transformer

for epoch in range(epochs):
    # ==========================================
    # PHASE A: TRAINING
    # ==========================================
    transformer_model.train()
    epoch_train_loss, correct_train, total_train = 0, 0, 0
    
    for texts, labels in train_loader:
        texts = texts.to(device)
        labels = labels.to(torch.long).to(device) # Prevent scalar float error
        
        optimizer.zero_grad()
        predictions = transformer_model(texts)
        loss = criterion(predictions, labels)
        
        loss.backward()
        optimizer.step()
        
        epoch_train_loss += loss.item()
        _, predicted = torch.max(predictions, 1)
        total_train += labels.size(0)
        correct_train += (predicted == labels).sum().item()
        
    t_train_losses.append(epoch_train_loss / len(train_loader))
    t_train_accs.append(correct_train / total_train)
    
    # ==========================================
    # PHASE B: VALIDATION
    # ==========================================
    transformer_model.eval()
    epoch_val_loss, correct_val, total_val = 0, 0, 0
    
    with torch.no_grad():
        for texts, labels in val_loader:
            texts = texts.to(device)
            labels = labels.to(torch.long).to(device)
            
            predictions = transformer_model(texts)
            loss = criterion(predictions, labels)
            
            epoch_val_loss += loss.item()
            _, predicted = torch.max(predictions, 1)
            total_val += labels.size(0)
            correct_val += (predicted == labels).sum().item()
            
    t_val_losses.append(epoch_val_loss / len(val_loader))
    val_acc = correct_val / total_val
    t_val_accs.append(val_acc)
    
    if val_acc > t_best_val_acc:
        t_best_val_acc = val_acc
        
    print(f"Epoch {epoch+1}/{epochs} | "
          f"Train Loss: {t_train_losses[-1]:.4f}, Train Acc: {t_train_accs[-1]:.4f} | "
          f"Val Loss: {t_val_losses[-1]:.4f}, Val Acc: {t_val_accs[-1]:.4f}")

t_total_time = time.time() - t_start_time

print(f"\n--- Transformer Execution Report ---")
print(f"Total Transformer Training Time: {t_total_time:.2f} seconds")
print(f"Best Transformer Validation Accuracy: {t_best_val_acc * 100:.2f}%")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 5))

# Plot 1: Loss
plt.subplot(1, 2, 1)
plt.plot(range(1, epochs+1), t_train_losses, label='Transformer Train Loss', marker='o')
plt.plot(range(1, epochs+1), t_val_losses, label='Transformer Val Loss', marker='o')
plt.title('Transformer Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

# Plot 2: Accuracy
plt.subplot(1, 2, 2)
plt.plot(range(1, epochs+1), t_train_accs, label='Transformer Train Acc', marker='s')
plt.plot(range(1, epochs+1), t_val_accs, label='Transformer Val Acc', marker='s')
plt.title('Transformer Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

#### Strategic Comparison and Insights

The Transformer model converged significantly faster per epoch, climbing to an impressive 73.50% validation accuracy by just the second epoch, whereas the LSTM failed to break past a random-guess baseline (~53%) throughout the entire training process. Ultimately, the Transformer finished with a drastically higher best validation accuracy of 77.30%, proving it is far more capable of capturing text sentiments than the simple LSTM. Interestingly, the Transformer's parallel design did not translate into a shorter execution speed on this setup, taking 152.99 seconds compared to the LSTM's 130.11 seconds. This indicates that for a relatively small dataset and sequence length, the architectural overhead of multi-head self-attention and dual encoder layers outweighs the matrix-parallelization advantages offered by the hardware.
